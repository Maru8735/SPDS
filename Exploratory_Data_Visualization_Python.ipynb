```python
# Dictionary where keys are categories and values represent frequency
data = {'A': 3, 'B': 5, 'C': 2}

# Loop through each item and print blocks as symbols
# Each block represents one unit
for category, count in data.items():
    print(category, '⬛' * count)
```

```python
import matplotlib.pyplot as plt

# Categories and their corresponding values
categories = ['A', 'B', 'C']
values = [10, 15, 7]

# Create vertical bar chart
plt.bar(categories, values)

# Add title for clarity
plt.title("Bar Chart")

# Display the graph
plt.show()
```

```python
# Create pie chart using same values
plt.pie(values, labels=categories, autopct='%1.1f%%')

# autopct shows percentage on slices
plt.title("Pie Chart")

plt.show()
```

```python
import numpy as np

# Generate random data following normal distribution
data = np.random.randn(1000)

# Plot histogram with 30 class intervals (bins)
plt.hist(data, bins=30)

plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

```python
# Independent variable
x = [1, 2, 3, 4, 5]

# Dependent variable
y = [2, 4, 6, 8, 10]

# Plot line graph
plt.plot(x, y)

plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
# Compute frequency and bin edges
counts, bins = np.histogram(data, bins=10)

# Calculate midpoints of each class interval
midpoints = (bins[:-1] + bins[1:]) / 2

# Plot frequency polygon
plt.plot(midpoints, counts)

plt.title("Frequency Polygon")
plt.xlabel("Class Midpoints")
plt.ylabel("Frequency")

plt.show()
```

```python
from scipy.stats import norm

# Generate smooth x-values
x = np.linspace(min(data), max(data), 100)

# Plot normal distribution curve using mean and std deviation
plt.plot(x, norm.pdf(x, np.mean(data), np.std(data)))

plt.title("Frequency Curve")
plt.xlabel("Values")
plt.ylabel("Probability Density")

plt.show()
```

```python
# Calculate frequency distribution
counts, bins = np.histogram(data, bins=10)

# Compute cumulative frequency
cum_freq = np.cumsum(counts)

# Plot cumulative frequency against upper class boundary
plt.plot(bins[1:], cum_freq)

plt.title("Ogive Curve")
plt.xlabel("Upper Class Boundary")
plt.ylabel("Cumulative Frequency")

plt.show()
```

```python
# Generate random x and y values
x = np.random.rand(50)
y = np.random.rand(50)

# Plot scatter diagram
plt.scatter(x, y)

plt.title("Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")

plt.show()
```

```python
# Task names
tasks = ['Task A', 'Task B', 'Task C']

# Start times of tasks
start = [1, 3, 5]

# Duration of each task
duration = [2, 2, 3]

# Horizontal bar chart for scheduling
plt.barh(tasks, duration, left=start)

plt.title("Gantt Chart")
plt.xlabel("Time")

plt.show()
```

```python
import seaborn as sns

# Generate random 5x5 matrix
matrix = np.random.rand(5, 5)

# Plot heat map with values annotated
sns.heatmap(matrix, annot=True)

plt.title("Heat Map")
plt.show()
```

```python
# Box plot shows median, quartiles, and outliers
plt.boxplot(data)

plt.title("Box-Whisker Plot")
plt.ylabel("Values")

plt.show()
```

```python
# X-axis values
x = [1, 2, 3, 4]

# Y-axis values
y = [3, 5, 2, 6]

# Fill area under the curve
plt.fill_between(x, y)

plt.title("Area Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
import pandas as pd

# Create DataFrame with random values
df = pd.DataFrame(
    np.random.rand(100, 4),
    columns=['A', 'B', 'C', 'D']
)

# Compute correlation between variables
corr = df.corr()

# Visualize correlation matrix
sns.heatmap(corr, annot=True)

plt.title("Correlation Matrix")
plt.show()
```
